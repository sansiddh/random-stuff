```python
df_a = pd.read_csv(join(ROOT_DIR, 'epfo-35+.csv'))
df_b = pd.read_csv(join(ROOT_DIR, 'epfo-29-35.csv'))
df_c = pd.read_csv(join(ROOT_DIR, 'epfo-26-28.csv'))
df_d = pd.read_csv(join(ROOT_DIR, 'epfo-22-25.csv'))
df_e = pd.read_csv(join(ROOT_DIR, 'epfo-18-21.csv'))
df_f = pd.read_csv(join(ROOT_DIR, 'epfo--18.csv'))

for df in [df_a, df_b, df_c, df_d, df_e, df_f]:
  df.set_index('State', inplace=True)
  df.columns = [x.strip() for x in df.columns]
  df = df[[x for x in df.columns if len(x) > 0]]
```

```python
df_total = pd.concat([df_a, df_b, df_c, df_d, df_e, df_f])
df_total.drop([''], axis=1, inplace=True)
df_total
```

```python
df_total = df_total.groupby(df_total.index).sum()
df_total = df_total.sort_values('2022-23', ascending=False)
df_total['H1 2023-24'] = df_total[['Apr-23', 'May-23', 'Jun-23', 'Jul-23', 'Aug-23', 'Sep-23']].sum(axis=1)
```
